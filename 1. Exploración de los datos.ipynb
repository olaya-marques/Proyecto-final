## 1. Antecedentes del tema üìñ

Cada a√±o todas las empresas contratan a nuevos empleados, para lo cual, invierten gran cantidad de tiempo y dinero para la formaci√≥n y capacitaci√≥n de los mismos. Adem√°s, de la necesidad de mantener el talento (employee engagement) de los ya existentes (aparte de las nuevas incorporaciones); y es en este √°rea d√≥nde Recursos Humanos o de lo que  de aqu√≠ en adelante trataremos como **"People Analytics"** toma gran importancia.

Toda clave de √©xito empresarial reside en las personas que conforman la empresa es por ello que dedicar parte de los recursos en conocer a la plantilla es una necesidad real. People Analytics se define como un m√©todo de investigaci√≥n basado en datos cuyo objetivo es estudiar las personas que constituyen y forman parte de la entidad, de tal modo que el an√°lisis inteligente de dichos datos extra√≠dos, permita establecer conclusiones fiables y v√°lidas para la toma de decisiones por parte de los altos directivos en relaci√≥n, a la organizaci√≥n y estructura de la plantilla de su empresa.

### 1.1 Usos de People Analytics üë•

Dentro del departamento de Recursos Humanos, los usos de est√° nueva tecnolog√≠a son tan variados como las preguntas que queramos realizar o los aspectos a investigar sobre tales empleados. Un estudio realizado por Deloitte destaca las siguientes 5 funciones:

- **Reclutamiento** : selecci√≥n del talento mediante la comparaci√≥n de los curr√≠culums anteriores y similares  examinando su progresi√≥n laboral del puesto en cuesti√≥n dentro de la empresa.
- **Desarrollo**: establecer el mejor modelo de formaci√≥n para cada empleado (educaci√≥n personalizada).
- **Lealtad**: identificar en tiempo real empleados con signos de desmotivaci√≥n o intenciones de abandonar la empresa (Attrition).
- **Compensaci√≥n**: entender la cantidad e importancia de la remuneraci√≥n. Construcci√≥n de modelos matem√°ticos que permitan predecir el impacto de las modificaciones salariales.
- **Organizaci√≥n**: comprensi√≥n de las relaciones laborales (independientemente del nivel jer√°rquico) mediante el an√°lisis del flujo de mensajes, correos u otros medios de comunicaci√≥n entre el emisor y el receptor.

En definitiva, son muchas las funciones reales aplicables que generan grandes beneficios y ventajas competitivas para la empresa:

- Mejorando la gesti√≥n de las personas.
- Seguimiento y evaluaci√≥n de eficacia de las decisiones estrat√©tigas tomadas.
- Identificaci√≥n de nuevas √°reas de mejora.
- Distribuci√≥n eficiente de los recursos de la empresa (reduciendo costes de tiempo y dinero).
- Aumento del bienestar y satisfacci√≥n de los empleados y en consecuencia incrementa los √≠ndices de productividad.

*üìå De entre todas las funciones o usos de esta tecnol√≥gia en esta investigaci√≥n se tratar√° en profundidad la lealtad y la retenci√≥n de talento o employee engagement.*

### 1.2 Attrition en People Analytics üèÉüèΩ

**"Attrition"**, en espa√±ol "desgaste" se refiere a la p√©rdida de empleados a lo largo de un tiempo.

La deserci√≥n relativamente alta puede suponer un problema para las empresas: las ofertas de trabajo, los procesos de contrataci√≥n, el papeleo y la formaci√≥n de los nuevos empleados suponen grandes costes, que las empresas han de asumir frente a la p√©rdida de empleados. Adem√°s, la continua rotaci√≥n impide a las organizaciones el crecimiento en conocimientos y experiencia colectiva a largo plazo. Siendo especialmente preocupante en organizaciones o empresas de servicio orientadas al cliente, ya que se prioriza la interacci√≥n con personas conocidas y dichas rotaciones pueden generarles falta de confianza y por ende de lealtad.

A pesar de que los profesionales de Recursos Humanos toman medidas y asumen el liderazgo en el dise√±o de programas de compensaci√≥n y motivaci√≥n para la retenci√≥n de estos, en muchas ocasiones no resulta suficiente. 

üìå Es por ello que mi estudio se basar√° en el an√°lisis de las variables que m√°s afectan a los empleados y que pueden fomentar est√° deserci√≥n laboral tan com√∫n en las organizaciones. A prop√≥sito de elaborar un modelo predictivo capaz de analizar las posibilidades y predecir si el empleado o la plantilla en cuesti√≥n tiene intenciones o se encuentra en en el umbral de decisi√≥n de poder abandonar la entidad.
